# **apply**

creating a new feature by using an old one

```python
Df['female'].head()
```

```python
def gender_id_to_string (gender_id):
       return 'female' if gender_id == 1 else 'male'
```

```python
Df['female'].apply(gender_id_to_string)
```

**lamda with apply**

```python
Df['female'].apply(lambda gender_id :
                  'female' if gender_id == 1 else 'male')
```

**the same can be done with map method**

```python
Df['female'].map({0:'male', 1:'female'}).head()
```
